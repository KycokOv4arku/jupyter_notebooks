```python
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
def apply_darker_theme():
    plt.style.use("default")  # Reset to default first

    plt.rcParams.update(
        {
            # Remove unnecessary spines and ticks
            "axes.spines.top": False,
            "axes.spines.right": False,
            "xtick.minor.visible": False,
            "xtick.major.size": 0,
            "ytick.major.size": 0,
            "ytick.minor.visible": False,
            # Title size and padding
            "axes.titlesize": 20,  # Title font size
            "axes.titlepad": 20,  # Title padding from the graph
            # X and Y label sizes
            "axes.labelsize": 14,  # X and Y label font size
            # Text on each bar size
            "font.size": 12,  # General text size (includes bar labels)
            # Line styles
            "scatter.marker": "H",
            "lines.linewidth": 2.5,
            "yaxis.labellocation": "center",
            "xaxis.labellocation": "center",
            "axes.linewidth": 1.5,
            # Colors
            "text.color": "#fff",
            "axes.facecolor": "#282828",
            "figure.facecolor": "#282828",
            "axes.labelcolor": "#eeeeee",
            "axes.edgecolor": "#eeeeee",
            "xtick.color": "#eeeeee",
            "ytick.color": "#eeeeee",
            # Color cycle
            "axes.prop_cycle": cycler(
                "color",
                ["#FED789", "#A4BED5", "#72874E", "#023743", "#476F84", "#453947"],
            ),
            #
            "font.family": "Calibri",
        }
    )


apply_darker_theme()
```

```python
# hrs per day data points. For now adding them manually.
study = {
    "Jan'24": 51 / 30,
    "Feb'24": 43 / 29,
    "Mar'24": 21 / 31,
    "Apr'24": 24 / 30,
    "May'24": 16 / 31,
    "Jun'24": 55 / 30,
    "Jul'24": 72 / 31,
    "Aug'24": 87 / 31,
    "Sep'24": 75 / 30,
    "Oct'24": 118 / 31,
    "Nov'24": 117 / 30,
    "Dec'24": 38 / 31,
    "Jan'25": 11.5 / 31,
    "Feb'25": 53 / 28,
}
```

```python
# Create bar chart
fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(study.keys(), study.values(), width=0.7, align="center")  # Save bars

# Add labels and title
ax.set_xlabel("Month")
ax.set_ylabel("Hours")
ax.set_title("Average Study Hours per Day")

# Add data labels on top of each bar
ax.bar_label(bars, fmt="%.1f", padding=2)

# Display the chart with tight layout
plt.tight_layout()
plt.show()
```
